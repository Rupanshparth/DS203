From the above cell I can know how many missing values are there in each column

```python
missing_values = [] # Creating a list
# Stores number of missing values in each column in the list missing_values
for col in data.columns.values: 
    missing_values.append(np.sum(pd.isnull(data[col])))
    
df = pd.DataFrame({'category': data.columns.values, 'missing_values': missing_values}) # Creates a dataframe with index of rows as column names in data and missing values of the corresponding index as row element
df.set_index('category', inplace = True)
df.sort_values('missing_values', inplace=True) # Sorts the elements in ascending order
df.plot(y='missing_values', kind='bar', legend=False) # plots in bar graph format
```

```python
df.sort_values('missing_values', ascending = False, inplace = True) # sorts in descending order
df.plot(y='missing_values', kind='bar', legend=False) # plots in bar graph format
```

The above two cells plot the histogram of number of missing values in each column in ascending and descending order.

```python
data.head() # First 5 rows of data for getting an idea to clean the data
# The country column is useless, all the entries are India
# The year column is also useless, since there is only a single unique value which is 2011
# Then there are columns which have many missing value
```

```python
# Dropping the Year and Country column

data = data.drop(['Country'], axis = 1)
data = data.drop(['Year'], axis = 1)

```

```python
data['Number of census houses'].head(50)
```

```python
c = data.columns.values
# Filling the missing values in object type number of houses columns with 0

data = data.replace(to_replace="zero",value = float(0))
#data[c[7]] = data[c[7]].fillna(float(0))
#data[c[8]] = data[c[8]].fillna(float(0))

# Changing the datatype of number of houses columns from object to float

data[c[6]] = pd.to_numeric(data[c[6]])
data[c[7]] = pd.to_numeric(data[c[7]])
data[c[8]] = pd.to_numeric(data[c[8]])
```

```python
col = list(data.columns) # getting all the columns in the form of list

# Keeping only the columns having object/mixed values 

for c in data.columns.values: 
    if (data[c].dtypes != 'object'):
        col.remove(c)



# Removing leading and trailing characters from columns with str type

for i in col:
    data[i] = data[i].str.strip()

# Filling the missing values in columns with datatype string
for c in data.columns.values: 
    if (data[c].dtypes == 'object'):
        data[c] = data[c].fillna("None") 
# Filling the missing values of Float types with the mean values (don't need to do anything about int types, have no missing values)
for c in data.columns.values: 
    if (data[c].dtypes == 'float64'):
        data[c] = data[c].fillna(np.mean(data[c])) 
```

The above three cells do data cleaning, the types of cleaning performed are explained using comments. Gives us the data in a form which allows us to extract information clearly

```python
# Checking if there are still any missing values
for col in data.columns.values:
    print(col + ' has ' + str(np.sum(pd.isnull(data[col]))) + ' missing values')
```

```python
# Histograms for discrete values
for col in data.columns.values: # for all columns
  # if the values are integers or discrete
  if (data[col].dtypes == 'int64') | (data[col].dtypes == 'object'):
    plt.hist(data[col])
    plt.xlabel(col)
    plt.show()
```

```python
# Plot histograms for continuous variables

for col in data.columns.values:
    if(data[col].dtypes == 'int64') | (data[col].dtypes == 'float64'):
        plt.hist(data[col],10 if len(pd.unique(data[col]))>20 else len(pd.unique(data[col])))
        plt.xlabel(col)
        plt.show()
```

```python
data_continuous = data 

# dropping all the discrete variables

for c in data.columns.values:
    if (data[c].dtypes == 'object'):
        data_continuous = data_continuous.drop([c], axis=1)

#data_continuous.dtypes
sns.pairplot(data_continuous) # Plots a scatter plot for each pair with permutations
plt.show() # Displays the plot
```

```python
corrMatrix = data_continuous.corr(method='spearman') # Use spearman for non-linear
fig, ax = plt.subplots(figsize=(20,16))
sns.heatmap(abs(corrMatrix), annot=False) # Show absolute value
plt.show()
```

The above heatmap is very helpful, it allows us to know how much one variable is related to other
